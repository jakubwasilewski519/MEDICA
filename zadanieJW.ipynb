```python
import random
import time

print("===================================")
print("        GRA: LOCH POTWORA")
print("===================================")

imie = input("Podaj imię bohatera: ")

zycie_gracza = 100
zycie_potwora = 120
mikstury = 3
zloto = 0

print()
print(f"Witaj, {imie}!")
print("Wchodzisz do ciemnego lochu...")
time.sleep(1)
print("Nagle z cienia wychodzi ogromny potwór!")
time.sleep(1)

while zycie_gracza > 0 and zycie_potwora > 0:
    print()
    print("----------- STATUS -----------")
    print(f"{imie} - życie: {zycie_gracza}")
    print(f"Potwór - życie: {zycie_potwora}")
    print(f"Mikstury leczenia: {mikstury}")
    print("------------------------------")

    print()
    print("Co chcesz zrobić?")
    print("1. Atak mieczem")
    print("2. Silny atak")
    print("3. Użyj mikstury leczenia")
    print("4. Próba ucieczki")

    wybor = input("Wybierz opcję 1-4: ")

    if wybor == "1":
        obrazenia = random.randint(12, 25)
        zycie_potwora -= obrazenia
        print(f"Atakujesz mieczem i zadajesz {obrazenia} obrażeń!")

    elif wybor == "2":
        szansa = random.randint(1, 100)

        if szansa <= 60:
            obrazenia = random.randint(25, 45)
            zycie_potwora -= obrazenia
            print(f"Silny atak trafia! Zadajesz {obrazenia} obrażeń!")
        else:
            print("Silny atak nie trafił!")

    elif wybor == "3":
        if mikstury > 0:
            leczenie = random.randint(25, 40)
            zycie_gracza += leczenie

            if zycie_gracza > 100:
                zycie_gracza = 100

            mikstury -= 1
            print(f"Używasz mikstury i odzyskujesz {leczenie} punktów życia.")
        else:
            print("Nie masz już mikstur!")

    elif wybor == "4":
        szansa_ucieczki = random.randint(1, 100)

        if szansa_ucieczki <= 35:
            print("Udało ci się uciec z lochu!")
            break
        else:
            print("Nie udało się uciec!")

    else:
        print("Nieprawidłowy wybór. Tracisz turę!")

    if zycie_potwora <= 0:
        break

    time.sleep(1)

    obrazenia_potwora = random.randint(10, 30)
    zycie_gracza -= obrazenia_potwora
    print(f"Potwór kontratakuje i zadaje ci {obrazenia_potwora} obrażeń!")

    time.sleep(1)

print()
print("===================================")

if zycie_gracza <= 0:
    print("Niestety, potwór cię pokonał...")
    print("KONIEC GRY")
elif zycie_potwora <= 0:
    zloto = random.randint(50, 150)
    print("Pokonałeś potwora!")
    print(f"Zdobywasz {zloto} sztuk złota.")
    print("ZWYCIĘSTWO!")
else:
    print("Opuściłeś loch bez walki do końca.")
    print("KONIEC GRY")

print("===================================")
```
